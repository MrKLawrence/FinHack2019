```python
import pandas as pd
import numpy as np
```

```python
FAV1 = 'Aveda'
#FAV2 = 'Optical 88'
```

```python
df = pd.read_csv('ratings.csv') #importing the shop ratings data
```

```python
shop_titles = pd.read_csv('shop_list.csv') #importing the shopID-title list
```

```python
df = pd.merge(df, shop_titles, on='shopID') #merging the two dataframe, showing the shop title according to its shopID
df.head(10)
```

```python
shop_matrix = df.pivot_table(index='user_id', columns='title', values='rating')
shop_matrix.head()
# converting the dataset into a matrix: shop titles as the columns; user_id as index; 
# ratings as values.
# NaN = didnt rate

# Matrix is built to compute the correlation between the ratings of a shop and the rest of the shops in the mall
# pandas pivot_table is used to create the shop matrix
```

```python
Fav1_user_rating = shop_matrix[FAV1]
#create dataframe with the ratings of these shops from our shop_matrix
```

```python
Fav1_user_rating.head()
```

```python
similar_to_fav1=shop_matrix.corrwith(Fav1_user_rating)
similar_to_fav1.head()
# pandas corrwith function is used to compute the correlation between two dataframe
# pairwise correlation of rows or columns of two dataframe objects
```

```python
corr_fav1 = pd.DataFrame(similar_to_fav1, columns=['Correlation'])
corr_fav1.dropna(inplace=True) #drop all null values and show all the related shops
corr_fav1.head()
```
